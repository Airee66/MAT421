## Section 22.1 - ODE Initial Value Problem Statement

An Ordinary Differential Equation (ODE) can be written as:
$$  F\left(x, f(x), \frac{df(x)}{dx}, \frac{d^2f(x)}{dx^2}, \frac{d^3f(x)}{dx^3}, \ldots, \frac{d^{n-1}f(x)}{dx^{n-1}}\right) = \frac{d^{n}f(x)}{dx^{n}} $$



ODEs solved using initial values to find a unique solution are called initial value problems.  These solutions can provide important insight on how a system changes over time.

## Section 22.2 - Reduction of Order

## Section 22.3 - The Euler Method

#### Explicit Euler Formula

Below is the initial value problem $ \frac {dy} {dt} = cos(t) $ with initial condition $y(0) = -1 $ solved using the Explicit Euler Method in Python.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: np.cos(t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (10, 4))
plt.plot(t, np.sin(t)-1, 'g', label='Exact')
plt.plot(t, s, 'b--', label='Approximate')
plt.title('Approximate and Exact Solution \
for cos(t), y(0) = -1')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

Increasing the number of steps (h) provides a better approximation:

```python
h = 0.01 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (10, 4))
plt.plot(t, np.sin(t)-1, 'g', label='Exact')
plt.plot(t, s, 'b--', label='Approximate')
plt.title('Approximate and Exact Solution \
for cos(t), y(0) = -1')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
