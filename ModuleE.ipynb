## 3.2 - Continunity and Differentiation

### 3.2.1 - Limits and Continuity
limits are an essential concept in calculus because they allow us to study the behavior of functions near a specific point, calculate derivatives and integrals, and define continuity.

##### Continuous Function
Let $f : D \to R$ be a real-valued function on $D \subseteq R^d $.  The $f$ is said to be continuous at $a \in D $ if

$ \\lim_{a\to\infty} f(x) = f(a)$


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,4*np.pi,0.1)
f = np.cos(x)

plt.plot(f)
plt.title('Continous Function cos(x)')
plt.xlabel('x')
plt.ylabel('y')
```

```python
from sympy import *

x = symbols('x')
f = cos(x)
limit_expr = limit(f, x, 0) 
print('limit for cos(x) as x approaches 0: ', limit_expr)
```

### 3.2.2 - Derivatives

Let $f: D \to R$ where $D\subseteq R$ and let $x_0 \in D$ be an interior point of $D$.  The derivative of $f$ at $x_0$ is 

$ \\f'(x_0) = \frac{df(x_0)}{dx} = \lim_{h \to 0} \frac{f(x_0 + h) - f(x_0)}{h}  $

provided the limit exists.

```python
print('derivative of cos(x) = ', diff(cos(x)))
print('derivative of x^2 + x - 9 = ', diff(x**2 + x - 9))
```

### 3.2.3 - Taylor’s Theorem

Taylor theorem is used to approximate the behavior of a function near a given point by computing its derivatives at that point and evaluating them at different powers of the distance from the point.

$ \\f(b) = f(a) + (b-a)f'(a) + \frac12 (b-a)^2 f''(a) + ... + \frac{(b-a)^{m-1}}{(m-1)!} f^{m-1}(a) + R_m $

```python
import math
%matplotlib inline

def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )
    
    return cos_approx

angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_cos = np.cos(angles)

#figure setup
fig, ax = plt.subplots()
ax.plot(angles,p_cos)

#series
for i in range(1,6):
    t_cos = [func_cos(angle,i) for angle in angles]
    ax.plot(angles,t_cos)

ax.set_ylim([-7,4])

#legend
legend_lst = ['cos(x) function']
for i in range(1,6):
    legend_lst.append(f'Taylor Series - {i} terms')
ax.legend(legend_lst, loc=3)

plt.show()
```

### 3.3.3 - Gradient Descent

Gradient descent is an iterative optimization algorithm for finding a local minimum of a differentiable function.

```python
cur_x = 3 
rate = 0.01 
precision = 0.000001 
previous_step_size = 1 
max = 10000 
iters = 0 
f = lambda x: x**3 + 6 

while previous_step_size > precision and iters < max:
    prev_x = cur_x 
    cur_x = cur_x - rate * f(prev_x) #Gradient descent
    previous_step_size = abs(cur_x - prev_x) 
    iters = iters+1 
    #print("Iteration",iters,"\nX value is",cur_x)
    
print("Local minimum occurs at", cur_x)

x = np.linspace(-5,5,20)
y = x**3 + 6

#function plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#local min
plt.plot(x,y, 'r')
#plt.plot(cur_x, 'b')
plt.plot(cur_x, 0, marker="o", markersize=10, markerfacecolor="green")
plt.show()
```
